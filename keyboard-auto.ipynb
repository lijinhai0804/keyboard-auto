```python
import keyboard
import time
import threading

class KeyboardListener:
    def __init__(self):
        self.is_running = False
        self.is_m_pressed = False
        self.c_thread = None
    
    def start(self):
        self.is_running = True
        threading.Thread(target=self._listen).start()
    
    def stop(self):
        self.is_running = False
        if self.c_thread:
            self.c_thread.stop()
        
    def _listen(self):
        while self.is_running:
            if keyboard.is_pressed("p"):
                if not self.is_m_pressed:
                    print("P键按下")
                    self.is_m_pressed = True
                    if self.c_thread:
                        self.c_thread.stop()
                        self.c_thread = None
                    else:
                        self.press_shift_w_and_hold_c()
            else:
                self.is_m_pressed = False
            time.sleep(0.1)
                
    def press_shift_w_and_hold_c(self):
        keyboard.press("shift")
        keyboard.press("w")
        self.c_thread = ContinuousKeyPressThread("c")
        self.c_thread.start()

class ContinuousKeyPressThread(threading.Thread):
    def __init__(self, key):
        super().__init__()
        self.key = key
        self.is_running = False
    
    def run(self):
        self.is_running = True
        while self.is_running:
            keyboard.press(self.key)
            time.sleep(0.1)
        keyboard.release(self.key)
    
    def stop(self):
        self.is_running = False

# 实例化键盘监听器并启动
keyboard_listener = KeyboardListener()
keyboard_listener.start()

# 等待程序结束
try:
    while True:
        time.sleep(1)
except KeyboardInterrupt:
    print("\n接收到键盘中断，程序即将退出...")
    keyboard_listener.stop()
    time.sleep(1)
    print("程序已退出")
```
