```python
import keyboard
import time
import threading

class KeyboardListener:
    def __init__(self):
        self.is_running = False
        self.shift_w_pressed = False
        self.c_thread = None
    
    def toggle(self):
        if self.is_running:
            self.stop()
        else:
            self.start()
    
    def start(self):
        if not self.is_running:
            self.is_running = True
            threading.Thread(target=self._listen).start()
    
    def stop(self):
        if self.is_running:
            self.is_running = False
            if self.shift_w_pressed:
                keyboard.release("shift")
                keyboard.release("w")
                self.shift_w_pressed = False
            if self.c_thread:
                self.c_thread.stop()
        
    def _listen(self):
        while self.is_running:
            if keyboard.is_pressed("p"):
                if not self.shift_w_pressed:
                    keyboard.press("shift")
                    keyboard.press("w")
                    self.shift_w_pressed = True
                    self.press_c_continuously()
                else:
                    self.stop()
            time.sleep(0.1)
    
    def press_c_continuously(self):
        self.c_thread = ContinuousKeyPressThread("c")
        self.c_thread.start()

class ContinuousKeyPressThread(threading.Thread):
    def __init__(self, key):
        super().__init__()
        self.key = key
        self.is_running = False
    
    def run(self):
        self.is_running = True
        while self.is_running:
            keyboard.press(self.key)
            keyboard.release(self.key)
            time.sleep(0.2)
    
    def stop(self):
        self.is_running = False

# 实例化键盘监听器
keyboard_listener = KeyboardListener()

# 等待P键开启/关闭代码运行
try:
    while True:
        if keyboard.is_pressed("p"):
            keyboard_listener.toggle()
            time.sleep(0.2)  # 防止重复触发
except KeyboardInterrupt:
    print("\n接收到键盘中断，程序即将退出...")
    keyboard_listener.stop()
    time.sleep(1)
    print("程序已退出")
```
